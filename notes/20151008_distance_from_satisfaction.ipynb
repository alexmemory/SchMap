```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
```

The distance from satisfaction for the rule

$b(x)\rightarrow i(y), c(x,y)$

is $\max(truth value of body - truth value of head, 0)$
where the truth value of the body is $b(x)$ (shorten to $b$).  The truth value of the head is 
$\max(c + i - 1, 0)$
where $c$ is $c(x,y)$ and $i$ is $i(y)$.  So

$d(i, b, c) = \max(b - \max(c + i - 1, 0), 0)$

```python
i = np.arange(0, 1, 0.05)
b = np.arange(0, 1, 0.05)
ii, bb = np.meshgrid(i, b)
for c in [0,.25,.5,1.]:
    fig = plt.figure()
    ax = fig.gca(projection = '3d')
    dist = np.vectorize(lambda i, b: max(b - max(c + i - 1, 0), 0))
    surf = ax.plot_surface(ii, bb, dist(ii,bb), rstride=1, cstride=1, 
                           cmap=cm.coolwarm, linewidth=0, antialiased=False)
    plt.xlabel('i'); plt.ylabel('b')
    plt.title('distance from satisfaction for c=%.02f'%c)
    plt.show()
```

At $c=1$ or $c=0$, distance from satisfaction is convex, while generally it is nonconvex.  For any c and while holding b constant, distance from satisfaction is quasiconvex.  E.g., for $c=.5$ and $b=.5$, distance from satisfaction is a constant non-minimal value .5 for values of $i$ between 0 and .5.  For values of i above .5, the function is convex and reaches the true minimum 0 at $i=1$.
